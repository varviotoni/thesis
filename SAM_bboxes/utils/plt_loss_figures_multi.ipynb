```python
import wandb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

api = wandb.Api()

# Define runs with different epoch lengths
runs_data = [
    {"run_id": "xt3bahti", "name": "5 epochs", "color": "blue"},
    {"run_id": "71nsiy3n", "name": "5 epochs", "color": "red"},
    {"run_id": "sp2bnpl8", "name": "5 epochs", "color": "green"}
]

# Function to process losses and normalize to 100 steps
def process_losses(run_id, target_length=100):
    run = api.run(f"tmv641-tampere-university/Thesis_datasplit_60_20_20_experimentation/{run_id}")
    history = run.scan_history()
    
    train_losses = [point['train_loss'] for point in history if 'train_loss' in point]
    val_losses = [point['validation_loss'] for point in history if 'validation_loss' in point]
    
    # Create pandas Series and drop NaN values
    train_series = pd.Series(train_losses).dropna()
    val_series = pd.Series(val_losses).dropna()
    
    train_losses_clean = train_series.values
    val_losses_clean = val_series.values
    
    # Calculate appropriate chunk sizes for normalization
    train_chunk_size = max(1, len(train_losses_clean) // target_length)
    val_chunk_size = max(1, len(val_losses_clean) // target_length)
    
    # Process losses - mean of chunks
    def chunk_mean(arr, chunk_size):
        return np.array([np.mean(arr[i:i+chunk_size]) for i in range(0, len(arr), chunk_size)])[:target_length]
    
    train_losses_mean = chunk_mean(train_losses_clean, train_chunk_size)
    val_losses_mean = chunk_mean(val_losses_clean, val_chunk_size)
    
    # Create x-axis values (epochs)
    epochs = np.arange(1,len(val_losses_mean)+1)
    
    return {
        "epochs": epochs,
        "train_losses": train_losses_mean,
        "val_losses": val_losses_mean
    }

# Create the plot
plt.figure(figsize=(5, 5))
plt.ylim(0, 5)

# Process and plot each run
for run_info in runs_data:
    processed_data = process_losses(run_info["run_id"])
    
    # Plot training loss with solid line
    plt.plot(processed_data["epochs"], processed_data["train_losses"], 
             color="blue", linestyle='-', 
             label=f'Training loss ({run_info["name"]})')
    
    # Plot validation loss with dashed line
    plt.plot(processed_data["epochs"], processed_data["val_losses"], 
             color="orange", linestyle='-', 
             label=f'Validation loss ({run_info["name"]})')

plt.xlabel('Epochs (%)')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Across Different Runs')
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
#plt.xticks(np.arange(0, 101, 10))  # 0-100% in steps of 10
#plt.yticks(np.arange(0, 5.5, 0.5))
plt.tight_layout()
#plt.savefig('multi_run_comparison.png', dpi=300)  # Save the figure
plt.show()
```

```python
import wandb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

api = wandb.Api()

# Define multiple runs with the same number of epochs (5 epochs in this example)
# You can replace these with your actual 5-epoch run IDs
five_epoch_runs = [
    {"run_id": "xt3bahti", "name": "Run 1", "line_style": "-"},
    {"run_id": "71nsiy3n", "name": "Run 2", "line_style": "--"},  # Replace with actual run ID
    {"run_id": "sp2bnpl8", "name": "Run 3", "line_style": "-."}   # Replace with actual run ID
]

# Setting up the plot
plt.figure(figsize=(5, 5))
plt.ylim(0, 4)

# Process each run
for i, run_info in enumerate(five_epoch_runs):
    # Get run data
    run = api.run(f"tmv641-tampere-university/Thesis_datasplit_60_20_20_experimentation/{run_info['run_id']}")
    history = run.scan_history()
    
    # Extract loss values
    train_losses = [point['train_loss'] for point in history if 'train_loss' in point]
    val_losses = [point['validation_loss'] for point in history if 'validation_loss' in point]
    
    # Clean data
    train_series = pd.Series(train_losses).dropna()
    val_series = pd.Series(val_losses).dropna()
    
    train_losses_clean = train_series.values
    val_losses_clean = val_series.values
    
    # Calculate means of chunks - adjust chunk sizes based on your data
    def chunk_mean(arr, chunk_size):
        return np.array([np.mean(arr[i:i+chunk_size]) for i in range(0, len(arr), chunk_size)])
    
    # For 5-epoch runs, adjust these chunk sizes based on your data points per epoch
    train_chunk_size = len(train_losses_clean) // 5  # Adjust to get 1 point per epoch
    val_chunk_size = len(val_losses_clean) // 5      # Adjust to get 1 point per epoch
    
    # Process losses
    train_losses_mean = chunk_mean(train_losses_clean, train_chunk_size)
    train_epochs_mean = np.arange(1, len(train_losses_mean)+1)
    
    val_losses_mean = chunk_mean(val_losses_clean, val_chunk_size)
    val_epochs_mean = np.arange(1, len(val_losses_mean)+1)
    
    # Use different colors for different runs
    color = plt.cm.viridis(i/len(five_epoch_runs))
    
    # Plot the data
    plt.plot(train_epochs_mean, train_losses_mean, 
             color="blue", linestyle=run_info["line_style"], 
             label=f'Training loss ({run_info["name"]})')
    
    plt.plot(val_epochs_mean, val_losses_mean, 
             color="orange", linestyle=run_info["line_style"], marker='o', 
             label=f'Validation Loss ({run_info["name"]})')

# Configure plot aesthetics
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Comparison of Multiple 5-Epoch Training Runs')
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Set the x-ticks to match the number of epochs
plt.xlim(1, 5)
plt.xticks(np.arange(0, 6.9, 1))  # Ticks for epochs 1-5
plt.yticks(np.arange(0, 5.1, 0.5))

plt.tight_layout()
#plt.savefig('multi_5epoch_comparison.png', dpi=300)
plt.show()
```
