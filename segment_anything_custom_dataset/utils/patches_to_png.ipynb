```python
from PIL import Image
import os

# Input directory containing TIF files
input_directory = "patches"

# Output directory for PNG files
output_directory = "patches_png"
os.makedirs(output_directory, exist_ok=True)

# Loop through TIF files in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".tif") or filename.endswith(".tiff"):
        tif_file = os.path.join(input_directory, filename)

        # Open the TIF file
        tif_image = Image.open(tif_file)

        # Remove the file extension and add .png
        png_filename = os.path.splitext(filename)[0] + ".png"

        # Save the image as a PNG file in the output directory
        png_file = os.path.join(output_directory, png_filename)
        tif_image.save(png_file)

        print(f"Converted {tif_file} to {png_file}")

print("Conversion complete.")
```
